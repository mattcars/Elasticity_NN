```python
import numpy as np
import matplotlib.pyplot as plt

import torch
import torch.nn as nn
from torch.utils.data import TensorDataset, Dataset, DataLoader
from torchvision import transforms, utils, datasets
from torch import optim as optim

from sklearn.model_selection import train_test_split

from tqdm import *

import os as os
```

# Import Data

```python
'''LOAD MATERIAL DATA FILE'''
def load_file(fileE, fileI):
    
    # unpack the data
    Enet = np.load(fileE)
    sample_ind = np.load(fileI, allow_pickle=True)
    
    # parse file name for data
    name = fileE.split('.')[0]
    
    # get discretization of material quadrant
    dx = int(name[8:10])
    
    # number of samples
    N = int(name[14:])
    
    # Retrun matrices
    sample = np.zeros((N, 2*dx, 2*dx))
    sample_base = np.zeros((N, dx, dx))
    
    for i in range(sample_ind.shape[0]):
        ind = sample_ind[i]
        
        base = np.zeros(dx**2)
        base[ind] = 1
        base1 = base.reshape((dx, dx))
        
        sample_base[2*i] = base1
        
        sample_base[2*i+1] = np.flip(np.rot90(base1), axis=0)
    
    return Enet, sample_base


'''BUILD FULL MATERIAL FROM BASE LAYOUT'''
def fill_base(base_array, E, v, typ='', norm=False):
    X = np.zeros((base_array.shape[0], 2, base_array.shape[1], base_array.shape[2]))
    
    # Fill in the binary data with E and v
    for i, base in enumerate(base_array):
        xE = np.zeros(base.shape)
        
        xE[base==1] = E[0]
        xE[base==0] = E[1]
        X[i, 0] = xE
        
        xE[base==1] = v[0]
        xE[base==0] = v[1]
        X[i, 1] = xE
    
    # Normalize by max E and v
    if norm:
        x = np.zeros(X.shape)
        x[:, 0, :, :] = X[:, 0, :, :]/max(E[0], E[1])
        x[:, 1, :, :] = X[:, 1, :, :]/max(v[0], v[1])
    else:
        x = X
    
    # Transform to Tensor
    if typ=='Tensor':
        x = torch.from_numpy(x).float()
    
    return x

'''CONSTRUCT THE FULL 10 X 10 MATERIAL SAMPLE'''
def construct_full(base):
    
    sample = np.zeros((base.shape[0], 10, 10))
    for i, base1 in enumerate(base):
        base2 = np.flip(base1, axis=1)
        base3 = np.flip(base1, axis=0)
        base4 = np.flip(base2, axis=0)
        
        top = np.concatenate((base1, base2), axis=1)
        bott = np.concatenate((base3, base4), axis=1)
        layout = np.concatenate((top, bott), axis=0)
        
        sample[i] = layout
        
    return sample
```

```python
# import created data
fileE = 'train_E_05x05_10000.npy'
fileI = 'train_I_05x05_10000.npy'

# Get data from files
'''
E = the training labels (Enet of each material)
base = training data (5x5) (the base design of each sample)
'''
E_all, base_all = load_file(fileE, fileI)

# Material properties
E_arr = [1000, 100]
v_arr = [.4, .1]
```

```python
# Normalize labels
y = E_all.reshape((E_all.shape[0], 1))/max(E_arr)
y = torch.from_numpy(y).float()

# Prepare data
X = torch.from_numpy(base_all).float()
noise = torch.randn(X.size()).float() * .06

X = X + noise
X = (X - X.min())
X = X/X.max()

# Compile as dataset
dataset = TensorDataset(X,y)

# See distorted images
i=804
fig, ax = plt.subplots(1, 2, figsize=(8, 8))
ax[0].imshow(X[i])
ax[1].imshow(base_all[i])
ax[0].set_title('Sample with added Noise')
ax[1].set_title('Original Sample')
```

# Network Architecture

```python
class Regression(nn.Module):
    
    # Network consruction
    def __init__(self, in_channels, ngpu=1):
        super(Regression, self).__init__()
        self.ngpu = ngpu
        
        # Convolutional 2D layers
        self.conv = nn.Sequential(
            nn.Conv2d(in_channels, out_channels=10, kernel_size=3, padding=1), # 2 x 5 x 5
            nn.ReLU(),
            nn.BatchNorm2d(10), 
            nn.MaxPool2d(kernel_size=3, padding=1, stride=1),
            nn.ReLU(),
            nn.BatchNorm2d(10),
            nn.Conv2d(10, 25, kernel_size=3), 
            nn.ReLU(),
            nn.BatchNorm2d(25),
            nn.Conv2d(25, 50, kernel_size=3),
            nn.ReLU(),
            nn.BatchNorm2d(50)
        )
        
        # Linear layers
        self.lin = nn.Sequential(
            nn.Linear(50, 20), 
            nn.ReLU(),
            nn.BatchNorm1d(20), 
            nn.Dropout(p=0.5),
            nn.Linear(20, 4),
            nn.ReLU(),
            nn.Linear(4, 1),
            nn.ReLU()
        )
       
        
    def forward(self, x):
        x = self.conv(x)
        x = x.reshape((x.shape[0], x.shape[1]*x.shape[2]*x.shape[3]))
        x = self.lin(x)
        return x
```

```python
class Generator(nn.Module):
    def __init__(self, z_size, device, ngpu=1):
        super(Generator, self).__init__()
        self.ngpu = ngpu
        self.device=device
        
        embed_len = 20
        input_dim = z_size + embed_len + 1
        output_dim = 5*5
        
        #self.embed = nn.Embedding(11, embed_len)
        self.embed = nn.Sequential(
            nn.Linear(1, embed_len)
        )
        
        self.lin = nn.Sequential(
            nn.Linear(input_dim, 10),
            nn.LeakyReLU(0.2), 
            nn.Linear(10, 15),
            nn.LeakyReLU(0.2),
            nn.Linear(15, 20),
            nn.LeakyReLU(0.2),
            nn.Linear(20, output_dim),
            nn.Sigmoid()
        )
        
    '''
    FORWARD FUNCTION
    z = noise vector (B x z_size)
    y = label vector (B x 1)
    '''
    def forward(self, z, y):
        # Create int representation of label (0-10)
        #y_int = ((abs(y)*10).int()).long()
        y_int = y
        y_int = self.embed(y_int).squeeze()
        
        x = torch.cat((z, y_int), dim=1)
        x = torch.cat((x, y), dim=1)
        
        x = self.lin(x)
        
        x = x.reshape(x.shape[0], 5, 5)
        
        return x.to(self.device)
```

```python
class Discriminator(nn.Module):
    def __init__(self, device, ngpu=1):
        super(Discriminator, self).__init__()
        self.ngpu = ngpu
        self.device=device
        
        input_dim = 5*5 + 9 + 1
        
        self.embed = nn.Embedding(11, 9)

        self.lin = nn.Sequential(
            nn.Linear(input_dim, 20),
            nn.LeakyReLU(0.2),
            nn.Dropout(0.2),
            nn.Linear(20, 10),
            nn.LeakyReLU(0.2),
            nn.Linear(10, 1),
            nn.Sigmoid()
        )
        
        self.lin0 = nn.Sequential(
            nn.Linear(input_dim, 30),
            nn.LeakyReLU(0.2),
            nn.Dropout(0.3),
            nn.Linear(30, 25),
            nn.LeakyReLU(0.2),
            nn.Dropout(0.3),
            nn.Linear(25, 20),
            nn.LeakyReLU(0.2),
            nn.Dropout(0.3),
            nn.Linear(20, 1),
            nn.Sigmoid()
        )
        
    '''
    FORWARD FUNCTION
    x = material microstructer vector (B x 5 x 5)
    y = label vector (B x 1)
    '''
    def forward(self, x, y):
        # Create int representation of label (0-10)
        y_int = ((abs(y)*10).int()).long()
        y_int = self.embed(y_int).squeeze()
        
        # Reshape material coming in
        x = x.reshape(x.shape[0], 25)
        
        # Create linear vector
        x = torch.cat((x, y_int), dim=1)
        x = torch.cat((x, y), dim=1)

        x = self.lin(x)
        
        return x.to(self.device)
```

```python
# custom weights initialization called on netG and netD
def weights_init(m):
    classname = m.__class__.__name__
    if classname.find('Conv') != -1:
        nn.init.normal_(m.weight.data, 0.0, 0.02)
    elif classname.find('BatchNorm') != -1:
        nn.init.normal_(m.weight.data, 1.0, 0.02)
        nn.init.constant_(m.bias.data, 0)
```

```python
z_size = 100

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Initialize GAN networks
disc = Discriminator(device)
gen = Generator(z_size, device)
disc.to(device)
gen.to(device)


# Path for storing networks
model_store_path = os.getcwd()+'/GAN_Models/'
date = 0
G_name = str(date) + '_Gen.pt'
D_name = str(date) + '_Disc.pt'

# Initialize weights of the GAN
preload = 0
if preload:
    gen.load_state_dict(torch.load(model_store_path + G_name))
    disc.load_state_dict(torch.load(model_store_path + D_name))
else:
    gen.apply(weights_init)
    disc.apply(weights_init)
    
    
# Regression Network for testing samples
reg = Regression(2)
reg.load_state_dict(torch.load(model_store_path+'/Reg.pt'))
```

# Train Model

```python
# Training Functions
loss_fn = nn.BCELoss()
learning_rate = 0.0002
optimizerD = optim.Adam(disc.parameters(), lr=learning_rate)
optimizerG = optim.Adam(gen.parameters(), lr=learning_rate)

# Batch
B = 100
dataloader = torch.utils.data.DataLoader(dataset, batch_size=B, shuffle=True)
```

```python
# Flag whether to train or not
train=1
# check if GAN is set to preload or not

if train:
    epochs = 50000
    it = []
    errorG = []
    errorD = []
    errorD_fake=[]
    errorD_real=[]
    acc_real = []
    acc_fake = []


    for i in tqdm(range(epochs)):

        for data in dataloader:
            # Unpack data from dataloader
            x_real = data[0]
            E_real = data[1]


            '''TRAIN DISCRIMINATOR ON REAL'''
            optimizerD.zero_grad()

            y_predD_real = disc(x_real, E_real)
            y_targD = torch.full((B,1), 1)

            lossD_real = loss_fn(y_predD_real, y_targD)
            lossD_real.backward()


            '''TRAIN DISCRIMINATOR ON FAKE'''
            z = torch.randn(E_real.shape[0], z_size).float()
            E_fake = abs(torch.randn(B, 1))
            E_fake = E_fake/E_fake.max()
            x_fake = gen(z, E_real)

            y_predD_fake = disc(x_fake, E_fake)
            y_targD = torch.full((B,1), 0)

            lossD_fake = loss_fn(y_predD_fake, y_targD)
            lossD_fake.backward(retain_graph=True)

            lossD = (lossD_real + lossD_fake)/2
            optimizerD.step()


            '''TRAIN GENERATOR'''
            optimizerG.zero_grad()

            y_predG = disc(x_fake, E_fake)
            y_targG = torch.full((B,1), 1)

            lossG = loss_fn(y_predG, y_targG)
            lossG.backward()

            optimizerG.step()



        if i%10==9 or i==0:
            it.append(i)
            errorD.append(lossD.item())
            errorG.append(lossG.item())
            errorD_fake.append(lossD_fake.item())
            errorD_real.append(lossD_real.item())

            a_real = (y_predD_real>=0.5).sum().float()/B
            a_fake = (y_predD_fake<0.5).sum().float()/B

            acc_real.append(a_real)
            acc_fake.append(a_fake)

        if i%50==49:
            torch.save(gen.state_dict(), model_store_path + G_name)
            torch.save(disc.state_dict(), model_store_path + D_name)
```

```python
# Plot error and accuracy over epochs
if train:
    fig, (ax0, ax1) = plt.subplots(1, 2, figsize=(8, 4))

    ax0.set_title('Error')
    dplot=ax0.plot(it, errorD)
    gplot=ax0.plot(it, errorG)
    drplot = ax0.plot(it, errorD_real)
    dfplot = ax0.plot(it, errorD_fake)
    ax0.legend([dplot[0], gplot[0], drplot[0], dfplot[0]], ['Disc', 'Gen', 'Disc Real', 'Disc Fake'])

    ax1.set_title('Accuracy')
    pfake=ax1.plot(it, acc_fake)
    preal=ax1.plot(it, acc_real)
    ax1.legend([pfake[0], preal[0]], ['Fake', 'Real'])
    ax1.set_ylim([0, 1])
```
